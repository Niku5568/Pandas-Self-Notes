# find ranking of each movies in the group according to IMDB rating

```python
def movies_rank(group):
    group["genre_rank"]= group["IMDB_Rating"].rank(ascending=False)
    return group
```

```python
genre.apply(movies_rank)
```

# find normalised IMDB rating group wise

```python
def normalised_rating(group):
    group["normal_rating"]=(group["IMDB_Rating"]-group["IMDB_Rating"].min()) / ( group["IMDB_Rating"].max() - group["IMDB_Rating"].min() )
    return group

genre.apply(normalised_rating)
```

# create a function that can return the highest score of the any batsman

```python
temp=ipl[ipl["batsman"]=="V Kohli"]
temp.groupby("match_id")["batsman_runs"].sum().sort_values(ascending=False).head(1).values[0]
```

```python
def high(batsman):
    temp=ipl[ipl["batsman"]==batsman]
    return temp.groupby("match_id")["batsman_runs"].sum().sort_values(ascending=False).head(1).values[0]

batsman=input("enter batsman name")
high(batsman)
```
