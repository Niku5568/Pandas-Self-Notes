12. info                                              : show all basic infromation as per column and memory used by dataset

16. drop_duplicates                                   : remove duplicate from dataset

18. dropna                                           : will remove all the rows have missing values
19. fillna(value)                                     : fill na with your given value
20. inplace=True                                      : make permanet change in dataset
21. rename(columns={"value name":"new value"})        : rename columns name
22. apply                                             : help to apply user_defined function in the dataset
23. contains                                          : check string value present in the string or dataset list or not
24. set_index                                         : will convert any column into your index.
25. value_counts                                      : count occurence of value/number in the dataset (if found duplicate)

33. get_group                                         : used to extract a group from a GroupBy object
34. first                                             : returns the first row of each group
35. last                                              : returns the last row of each group
36. nth(index number)                                 : returns the nth row of each group (0-indexed)

38. selected column doing agg                         : agg({

                                                       DataFrame Operations
--------------------------------------------------------------------------------------------------------------------------------------------------------

1.  fetch single columns                              : DataFrame name["column name"] -> movies["title_x"]

4.  series of row                                     : DataFrame name.iloc [0 : 5 ] 5 is row index number
5.  fetch multiple rows                               : DataFrame name.iloc[ [ row1,row5, row10 ] ]
6.  fetch row by label                                : DataFrame name.loc["row label name"]
7.  fetch multiple row by label                       : DataFrame name.loc[ ["label1 ", "label2" ] ]
8.  fetch row and columns both                        : DataFrame name.iloc [0 : 5, 0: 3]
                                                        (row start : row end : row gape, col start : col end : col gape)
