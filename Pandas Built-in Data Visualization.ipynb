# x and y ticks 

```python
x=[x for x in range(0,190,10)]
y=[x for x in range(0,10,3)]
print(x,"\n",y)
```

```python
data[["PetalLengthCm","PetalWidthCm"]].plot(title="IRIS Dataset",
                                            ylabel="Petal Length in Cm",
                                            xlabel="index",
                                            linestyle=":",
                                            xlim=(0,180),
                                            ylim=(0,10),
                                            xticks=x,
                                            yticks=y,
                                            figsize=(10,3)
                                           )
plt.show() 
```

# area Plot

```python
data[["PetalLengthCm","PetalWidthCm"]].plot(title="IRIS Dataset",
                                            ylabel="Petal Length in Cm",
                                            xlabel="index",
                                            linestyle=":",
                                            xlim=(0,180),
                                            ylim=(0,10),
                                            xticks=x,
                                            yticks=y,
                                            figsize=(10,3),
                                            kind="area"
                                           )
plt.show() 
```
