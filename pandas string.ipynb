# split -> get

```python
# seprate surname, name, and title from name columns
titanic["Last Name"]=titanic["Name"].str.strip().str.split(",").str.get(0)
```

```python
# n=1 1 for 1 space
titanic["Name"].str.strip().str.split(",").str.get(1).str.strip().str.split(" ",n=1,expand=True)
```

```python
titanic[["Title","First Name"]]=titanic["Name"].str.strip().str.split(",").str.get(1).str.strip().str.split(" ",n=1,expand=True)
```

```python
# search john both case lower and upper in name
titanic[titanic["First Name"].str.contains("john",case=False)]
titanic[titanic["First Name"].str.contains("john",case=False)].head(1)
```

# find lastname and check start and end with char vowels.

```python
# ^ -> start , $ -> ends  ("^[^aeiouAEIOU].+[^aeiouAEIOU]$") 

titanic[titanic["Last Name"].str.contains("^[^aeiouAEIOU].+[^aeiouAEIOU]$")] # not start and end with vowels  use ^ -> start and end with consonent.
titanic[titanic["Last Name"].str.contains("^[aeiouAEIOU].+[aeiouAEIOU]$")] # start and end with vowels
```
